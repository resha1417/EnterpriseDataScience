```python

N0=1000000 #max susceptible population
beta=0.4   # infection spread dynamics
gamma=0.1  # recovery rate


# condition I0+S0+R0=N0
I0=df_analyse.Germany[35]
S0=N0-I0
R0=0
```

```python
def SIR_model(SIR,beta,gamma):
    ''' Simple SIR model
        S: susceptible population
        I: infected people
        R: recovered people
        beta: 
        
        overall condition is that the sum of changes (differnces) sum up to 0
        dS+dI+dR=0
        S+I+R= N (constant size of population)
    
    '''
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          #S*I is the 
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return([dS_dt,dI_dt,dR_dt])
```

```python
SIR=np.array([S0,I0,R0])
SIR
```

```python
SIR=np.array([S0,I0,R0])
propagation_rates=pd.DataFrame(columns={'susceptible':S0,
                                        'infected':I0,
                                        'recoverd':R0})



for each_t in np.arange(100):
   
    new_delta_vec=SIR_model(SIR,beta,gamma)
   
    SIR=SIR+new_delta_vec
    
    propagation_rates=propagation_rates.append({'susceptible':SIR[0],
                                                'infected':SIR[1],
                                                'recovered':SIR[2]}, ignore_index=True)
```

```python
ydata = np.array(df_analyse.Germany[35:])
t=np.arange(len(ydata))
```

```python
print(ydata[0])
ydata[len(ydata)-1]
```

```python

# ensure re-initialization 
I0=ydata[0]
S0=N0-I0
R0=0
beta
```

```python

def SIR_model_t(SIR,t,beta,gamma):
    ''' Simple SIR model
        S: susceptible population
        t: time step, mandatory for integral.odeint
        I: infected people
        R: recovered people
        beta: 
        
        overall condition is that the sum of changes (differnces) sum up to 0
        dS+dI+dR=0
        S+I+R= N (constant size of population)
    
    '''
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          #S*I is the 
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return dS_dt,dI_dt,dR_dt
```

```python
def fit_odeint(x, beta, gamma):
    '''
    helper function for the integration
    '''
    return integrate.odeint(SIR_model_t, (S0, I0, R0), t, args=(beta, gamma))[:,1] 
# we only would like to get dI
```

```python
popt=[0.4,0.1]
fit_odeint(t, *popt)
```

```python
popt, pcov = optimize.curve_fit(fit_odeint, t, ydata)
perr = np.sqrt(np.diag(pcov))
    
print('standard deviation errors : ',str(perr), ' start infect:',ydata[0])
print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
```

```python
# get the final fitted curve
fitted=fit_odeint(t, *popt)
```

```python
plt.semilogy(t, ydata, 'o')   # Make a plot with log scaling on the y axis.
plt.semilogy(t, fitted)
plt.title("Fit of SIR model for Germany cases")
plt.ylabel("Population infected")
plt.xlabel("Days")
plt.show()
print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
print("Basic Reproduction Number R0 " , popt[0]/ popt[1])
print("This ratio is derived as the expected number of new infections (these new infections are sometimes called secondary infections from a single infection in a population where all subjects are susceptible. @wiki")
```

```python
t_initial=28
t_intro_measures=14
t_hold=21
t_relax=21

beta_max=0.4
beta_min=0.11
gamma=0.1
pd_beta=np.concatenate((np.array(t_initial*[beta_max]),
                       np.linspace(beta_max,beta_min,t_intro_measures),
                       np.array(t_hold*[beta_min]),
                        np.linspace(beta_min,beta_max,t_relax),
                       ))
```

```python
print(len(pd_beta))
pd_beta
```

```python

SIR=np.array([S0,I0,R0])
propagation_rates_2=pd.DataFrame(columns={'susceptible':S0,
                                        'infected':I0,
                                        'recoverd':R0})



for each_beta in pd_beta:
   
    new_delta_vec=SIR_model(SIR,each_beta,gamma)
   
    SIR=SIR+new_delta_vec
    
    propagation_rates_2=propagation_rates_2.append({'susceptible':SIR[0],
                                                'infected':SIR[1],
                                                'recovered':SIR[2]}, ignore_index=True)
```

```python
print(len(propagation_rates_2))
print(len(propagation_rates_2.infected))
propagation_rates_2
```

```python

fig, ax1 = plt.subplots(1, 1)

ax1.plot(propagation_rates_2.index,propagation_rates_2.infected,label='infected',linewidth=3)

t_phases=np.array([t_initial,t_intro_measures,t_hold,t_relax]).cumsum()
# ax1.bar(np.arange(len(propagation_rates.infected)),ydata, width=0.8,label=' current infected Germany',color='r')
ax1.bar(np.arange(len(ydata)),ydata, width=0.8,label=' current infected Germany',color='r')
ax1.axvspan(0,t_phases[0], facecolor='b', alpha=0.2,label='no measures')
ax1.axvspan(t_phases[0],t_phases[1], facecolor='g', alpha=0.3,label='hard measures introduced')
ax1.axvspan(t_phases[1],t_phases[2], facecolor='y', alpha=0.4,label='hold measures')
ax1.axvspan(t_phases[2],t_phases[3], facecolor='b', alpha=0.5,label='relax measures')
ax1.axvspan(t_phases[3],len(propagation_rates.infected), facecolor='g', alpha=0.6,label='repead hard measures')

ax1.set_ylim(10, 1.5*max(propagation_rates.infected))
ax1.set_yscale('log')
ax1.set_title('Szenario SIR simulations  (demonstration purposes only)',size=16)
ax1.set_xlabel('time in days',size=16)
ax1.legend(loc='best',
           prop={'size': 16});
```

```python
# %load src/visualization/visualize.py
import pandas as pd
import numpy as np

import dash
dash.__version__
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output,State

import plotly.graph_objects as go

import os
#print(os.getcwd())
#df_input_large=pd.read_csv('data/processed/COVID_final_set.csv',sep=';')


fig = go.Figure()

app = dash.Dash()
app.layout = html.Div([

    dcc.Markdown('''
    #  Applied Data Science on COVID-19 data

    Goal of the project is SIR analysis in 3 phases and understanding them.

    '''),

    dcc.Markdown('''
    ## SIR Analysis in 3 phases:
    ### 1st Phase: 
    This is the initial SIR analysis of Germany,where we have initialised the initial susceptibel population (N0) = 1000000, beta(infected rate)=0.4 and gamma(recovery rate)=0.1
    
    ### 2nd Phase: ToDo
    
    ### 3rd Phase: ToDo
    '''),


    dcc.Dropdown(
        id='choose methods',
        options=[ {'label': each,'value':each} for each in ['Phase1','Phase2','Phase3']],
        value=['Phase1','Phase2','Phase3'] # which are pre-sele
    ),


    dcc.Graph(figure=fig, id='main_window_slope')
])



@app.callback(
    Output('main_window_slope', 'figure'),
    [Input('choose methods', 'value'),
    ])
def update_figure(tab_list):
    
    traces = []
    if 'Phase1' in tab_list:
        traces.append(dict(x=propagation_rates.index,
                                    y=propagation_rates.infected,
                                    mode='markers+lines',
                                    opacity=0.9,
                                    name="Infected"
                                    ))

        traces.append(dict(x=propagation_rates.index,
                                    y=propagation_rates.recovered,
                                    mode='markers+lines',
                                    opacity=0.9,
                                       name="Recovered"
                                    )),
        traces.append(dict(x=propagation_rates.index,
                                    y=propagation_rates.susceptible,
                                    mode='markers+lines',
                                    opacity=0.9,
                           name="Susceptible"
                                    ))
        layout=dict (
                width=1280,
                height=720,
                xaxis={'title':'time in days',
                        'tickangle':-45,
                        'nticks':20,
                        'tickfont':dict(size=14,color="#7f7f7f"),
                      },
                yaxis={'title':'change'},
                title={'text':'DISPLAY ME!'},
        )
    elif 'Phase2' in tab_list:
        traces.append(dict(x=t,y=ydata,mode='markers',opcaity=0.9))
        traces.append(dict(x=t,y=fitted,mode='markers',opcaity=0.9))
        
        layout=dict (
                width=1280,
                height=720,
                xaxis={'title':'Days',
                        'tickangle':-45,
                        'tickfont':dict(size=14,color="#7f7f7f"),
                      },
                yaxis={'title':'Infected','type':'log'},
        )
        
    elif 'Phase3' in tab_list:
        traces.append(dict(x=propagation_rates_2.index,y=propagation_rates_2.infected,mode='line',opcaity=0.9,
                           name="infected"))
        t_phases=np.array([t_initial,t_intro_measures,t_hold,t_relax]).cumsum()
        print(t_phases)
        traces.append(dict(x=np.arange(len(ydata)),y=ydata,type='bar',opcaity=0.9,name="currently infected"))
        traces.append(dict(type='rect',xref='x',yref='y',
                           x0=0,y0=27,x1=40,y1=236122,
                           opcaity=0.9,fillcolor="LightSkyBlue",line_width=0.6,name='no measures'))
        layout=dict (
                width=1280,
                height=720,
                xaxis={'title':'Days',
                        'tickangle':-45,
                        'nticks':20,
                        'tickfont':dict(size=14,color="#7f7f7f"),
                      },
                yaxis={'title':'Infected','type':'log'},
#                 type='line',xref='x',yref='y', x0=10,x1=28,y0=27,y1=236122,
#                            opcaity=0.9,fillcolor="LightSkyBlue",line_width=0.6,name='no measures'
            
        )
        
    return {
            'data': traces,
            'layout':layout
    }
        

if __name__ == '__main__':

    app.run_server(debug=True, use_reloader=False)
```
